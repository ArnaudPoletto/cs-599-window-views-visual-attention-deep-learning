```python
import sys

sys.path.append("../")

import os
import numpy as np
import matplotlib.pyplot as plt

os.environ["KMP_DUPLICATE_LIB_OK"] = "True"

from src.config import PROCESSED_SALICON_PATH
from src.datasets.salicon_dataset import SaliconDataset
from src.utils.file import get_paths_recursive
```

```python
sample_folder_paths = get_paths_recursive(folder_path=PROCESSED_SALICON_PATH, match_pattern="*", file_type="d", recursive=False)
dataset = SaliconDataset(
    sample_folder_paths=sample_folder_paths,
    with_transforms=True,
)
```

```python
frame, ground_truths = dataset[0]
frame = frame.transpose(1, 2, 0)
frame = ((frame * 0.5 + 0.5) * 255).astype(np.uint8)

plt.imshow(frame)
plt.show()

n_ground_truths = ground_truths.shape[0]
plt.figure(figsize=(20, 10))
for i in range(0, n_ground_truths):
    plt.subplot(1, n_ground_truths, i + 1)
    plt.imshow(ground_truths[i, 0], cmap="gray")
plt.show()
```
